```python
# 문제1
# H0: (S기업의 점심 비용 평균은 6,300원이다) μ=6300
# H1: (S기업의 점심 비용 평균은 6,300원이 아니다) 𝜇≠6300

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats 
plt.rcParams['font.family'] ='D2coding'

data = pd.read_excel("C:/Users/darkr/OneDrive/Desktop/2학년1학기/확통/S기업_점심비용.xlsx")
mu=6300; n = 100
print(data.describe())

plt.hist(data, label='점심비용', bins=8)    
plt.legend(); plt.show()                                      

t = (np.mean(data) -mu) / (np.std(data, ddof=1) / np.sqrt(n))
print('t값=',t)
p = 1 -stats.t.cdf(t, n-1)
print('p값=',p)

result = stats.ttest_1samp(data.lunch, mu, alternative='two-sided')
print(result)

if p<0.05 : print('H0기각')
else : print('H0를기각할수없음')
    
```

```python
# 문제2
# H0: (파이썬 과목을 수강하는 대학생의 주당 스마트폰 이용 시간 평균은 35시간이다) μ=35
# H1: (파이썬 과목을 수강하는 대학생의 주당 스마트폰 이용 시간 평균은 35시간이 아니다) 𝜇≠35

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats 
plt.rcParams['font.family'] ='D2coding'

data = pd.read_csv("C:/Users/darkr/OneDrive/Desktop/2학년1학기/확통/스마트폰_이용시간.csv")
mu=35; n = 40
print(data.describe())

plt.hist(data, label='이용시간', bins=7)    
plt.legend(); plt.show()                                      

t = (np.mean(data) -mu) / (np.std(data, ddof=1) / np.sqrt(n))
print('t값=',t)
p = 1 -stats.t.cdf(t, n-1)
print('p값=',p)

result = stats.ttest_1samp(data, mu, alternative='two-sided')
print(result)

if p<0.05 : print('H0기각')
else : print('H0를기각할수없음')
    
```

```python
# 문제3
# H0: (소프트웨어 학과 학생들의 월 평균 스마트폰 이용 요금은 68,000원 이하이다) μ≤68000
# H1: (소프트웨어 학과 학생들의 월 평균 스마트폰 이용 요금은 68,000원보다 더 많다) μ>68000

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats 
plt.rcParams['font.family'] ='D2coding'

data = pd.read_csv("C:/Users/darkr/OneDrive/Desktop/2학년1학기/확통/스마트폰_이용요금.txt")
mu=68000; n = 80
print(data.describe())

plt.hist(data, label='이용요금', bins=8)    
plt.legend(); plt.show()                                      

t = (np.mean(data) -mu) / (np.std(data, ddof=1) / np.sqrt(n))
print('t값=',t)
p = 1 -stats.t.cdf(t, n-1)
print('p값=',p)

result = stats.ttest_1samp(data, mu, alternative='greater')
print(result)

if p<0.05 : print('H0기각')
else : print('H0를기각할수없음')
    
```

```python
# 문제4 이분산
# H0: (한국 공장과 중국 공장에서 생산한 제품의 평균 지속 시간에 차이가 없다) 𝜇한국 = 𝜇중국
# H1: (한국 공장과 중국 공장에서 생산한 제품의 평균 지속 시간에 차이가 있다) μ한국 = μ중국

import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt

data= pd.read_excel("C:/Users/darkr/OneDrive/Desktop/2학년1학기/확통/배터리,근로,성적,체중/배터리_지속시간.xlsx")
print(data.describe())
data.boxplot(column=['china', 'korea'], vert=False)
plt.show()

d1 = pd.read_excel("C:/Users/darkr/OneDrive/Desktop/2학년1학기/확통/배터리,근로,성적,체중/배터리_지속시간_china.xlsx")
d2 = pd.read_excel("C:/Users/darkr/OneDrive/Desktop/2학년1학기/확통/배터리,근로,성적,체중/배터리_지속시간_korea.xlsx")
result= stats.ttest_ind(d1.china, d2.korea, alternative='two-sided', equal_var=False)
print(result)
```

```python
# 문제5 이분산
# H0: (대전과 광주 지역의 상담원들의 주당 근로시간 평균은 같다) 𝜇대전 = 𝜇광주
# H1: (대전과 광주 지역의 상담원들의 주당 근로시간 평균은 다르다) μ대전 = μ광주

import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt

data= pd.read_excel("C:/Users/darkr/OneDrive/Desktop/2학년1학기/확통/배터리,근로,성적,체중/주당_근로시간.xlsx")
print(data.describe())
data.boxplot(column=['daejeon', 'gwangju'], vert=False)
plt.show()

result= stats.ttest_ind(data.daejeon, data.gwangju, alternative='two-sided', equal_var=False)
print(result)
```

```python
# 문제6
# H0: (기말시험 성적의 평균은 중간시험 성적의 평균과 같거나 더 낮다) μ기말≤μ중간
# H1: (기말시험 성적의 평균은 중간시험 성적의 평균보다 더 높다) μ기말>μ중간

import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt

data= pd.read_excel("C:/Users/darkr/OneDrive/Desktop/2학년1학기/확통/배터리,근로,성적,체중/확률및통계_성적.xlsx")
print(data[['midterm','final']].describe())
data.boxplot(column=['midterm','final'], vert=False)
plt.show()

result=stats.ttest_rel(data.midterm, data.final, alternative='greater')
print(result)                          
```
