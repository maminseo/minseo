```python
# Î¨∏Ï†ú1
# H0: (SÍ∏∞ÏóÖÏùò Ï†êÏã¨ ÎπÑÏö© ÌèâÍ∑†ÏùÄ 6,300ÏõêÏù¥Îã§) Œº=6300
# H1: (SÍ∏∞ÏóÖÏùò Ï†êÏã¨ ÎπÑÏö© ÌèâÍ∑†ÏùÄ 6,300ÏõêÏù¥ ÏïÑÎãàÎã§) ùúá‚â†6300

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats 
plt.rcParams['font.family'] ='D2coding'

data = pd.read_excel("C:/Users/darkr/OneDrive/Desktop/2ÌïôÎÖÑ1ÌïôÍ∏∞/ÌôïÌÜµ/SÍ∏∞ÏóÖ_Ï†êÏã¨ÎπÑÏö©.xlsx")
mu=6300; n = 100
print(data.describe())

plt.hist(data, label='Ï†êÏã¨ÎπÑÏö©', bins=8)    
plt.legend(); plt.show()                                      

t = (np.mean(data) -mu) / (np.std(data, ddof=1) / np.sqrt(n))
print('tÍ∞í=',t)
p = 1 -stats.t.cdf(t, n-1)
print('pÍ∞í=',p)

result = stats.ttest_1samp(data.lunch, mu, alternative='two-sided')
print(result)

if p<0.05 : print('H0Í∏∞Í∞Å')
else : print('H0Î•ºÍ∏∞Í∞ÅÌï†ÏàòÏóÜÏùå')
    
```

```python
# Î¨∏Ï†ú2
# H0: (ÌååÏù¥Ïç¨ Í≥ºÎ™©ÏùÑ ÏàòÍ∞ïÌïòÎäî ÎåÄÌïôÏÉùÏùò Ï£ºÎãπ Ïä§ÎßàÌä∏Ìè∞ Ïù¥Ïö© ÏãúÍ∞Ñ ÌèâÍ∑†ÏùÄ 35ÏãúÍ∞ÑÏù¥Îã§) Œº=35
# H1: (ÌååÏù¥Ïç¨ Í≥ºÎ™©ÏùÑ ÏàòÍ∞ïÌïòÎäî ÎåÄÌïôÏÉùÏùò Ï£ºÎãπ Ïä§ÎßàÌä∏Ìè∞ Ïù¥Ïö© ÏãúÍ∞Ñ ÌèâÍ∑†ÏùÄ 35ÏãúÍ∞ÑÏù¥ ÏïÑÎãàÎã§) ùúá‚â†35

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats 
plt.rcParams['font.family'] ='D2coding'

data = pd.read_csv("C:/Users/darkr/OneDrive/Desktop/2ÌïôÎÖÑ1ÌïôÍ∏∞/ÌôïÌÜµ/Ïä§ÎßàÌä∏Ìè∞_Ïù¥Ïö©ÏãúÍ∞Ñ.csv")
mu=35; n = 40
print(data.describe())

plt.hist(data, label='Ïù¥Ïö©ÏãúÍ∞Ñ', bins=7)    
plt.legend(); plt.show()                                      

t = (np.mean(data) -mu) / (np.std(data, ddof=1) / np.sqrt(n))
print('tÍ∞í=',t)
p = 1 -stats.t.cdf(t, n-1)
print('pÍ∞í=',p)

result = stats.ttest_1samp(data, mu, alternative='two-sided')
print(result)

if p<0.05 : print('H0Í∏∞Í∞Å')
else : print('H0Î•ºÍ∏∞Í∞ÅÌï†ÏàòÏóÜÏùå')
    
```

```python
# Î¨∏Ï†ú3
# H0: (ÏÜåÌîÑÌä∏Ïõ®Ïñ¥ ÌïôÍ≥º ÌïôÏÉùÎì§Ïùò Ïõî ÌèâÍ∑† Ïä§ÎßàÌä∏Ìè∞ Ïù¥Ïö© ÏöîÍ∏àÏùÄ 68,000Ïõê Ïù¥ÌïòÏù¥Îã§) Œº‚â§68000
# H1: (ÏÜåÌîÑÌä∏Ïõ®Ïñ¥ ÌïôÍ≥º ÌïôÏÉùÎì§Ïùò Ïõî ÌèâÍ∑† Ïä§ÎßàÌä∏Ìè∞ Ïù¥Ïö© ÏöîÍ∏àÏùÄ 68,000ÏõêÎ≥¥Îã§ Îçî ÎßéÎã§) Œº>68000

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats 
plt.rcParams['font.family'] ='D2coding'

data = pd.read_csv("C:/Users/darkr/OneDrive/Desktop/2ÌïôÎÖÑ1ÌïôÍ∏∞/ÌôïÌÜµ/Ïä§ÎßàÌä∏Ìè∞_Ïù¥Ïö©ÏöîÍ∏à.txt")
mu=68000; n = 80
print(data.describe())

plt.hist(data, label='Ïù¥Ïö©ÏöîÍ∏à', bins=8)    
plt.legend(); plt.show()                                      

t = (np.mean(data) -mu) / (np.std(data, ddof=1) / np.sqrt(n))
print('tÍ∞í=',t)
p = 1 -stats.t.cdf(t, n-1)
print('pÍ∞í=',p)

result = stats.ttest_1samp(data, mu, alternative='greater')
print(result)

if p<0.05 : print('H0Í∏∞Í∞Å')
else : print('H0Î•ºÍ∏∞Í∞ÅÌï†ÏàòÏóÜÏùå')
    
```

```python
# Î¨∏Ï†ú5 Ïù¥Î∂ÑÏÇ∞
# H0: (ÎåÄÏ†ÑÍ≥º Í¥ëÏ£º ÏßÄÏó≠Ïùò ÏÉÅÎã¥ÏõêÎì§Ïùò Ï£ºÎãπ Í∑ºÎ°úÏãúÍ∞Ñ ÌèâÍ∑†ÏùÄ Í∞ôÎã§) ùúáÎåÄÏ†Ñ = ùúáÍ¥ëÏ£º
# H1: (ÎåÄÏ†ÑÍ≥º Í¥ëÏ£º ÏßÄÏó≠Ïùò ÏÉÅÎã¥ÏõêÎì§Ïùò Ï£ºÎãπ Í∑ºÎ°úÏãúÍ∞Ñ ÌèâÍ∑†ÏùÄ Îã§Î•¥Îã§) ŒºÎåÄÏ†Ñ = ŒºÍ¥ëÏ£º

import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt

data= pd.read_excel("C:/Users/darkr/OneDrive/Desktop/2ÌïôÎÖÑ1ÌïôÍ∏∞/ÌôïÌÜµ/Î∞∞ÌÑ∞Î¶¨,Í∑ºÎ°ú,ÏÑ±Ï†Å,Ï≤¥Ï§ë/Ï£ºÎãπ_Í∑ºÎ°úÏãúÍ∞Ñ.xlsx")
print(data.describe())
data.boxplot(column=['daejeon', 'gwangju'], vert=False)
plt.show()

result= stats.ttest_ind(data.daejeon, data.gwangju, alternative='two-sided', equal_var=False)
print(result)
```

```python
# Î¨∏Ï†ú6
# H0: (Í∏∞ÎßêÏãúÌóò ÏÑ±Ï†ÅÏùò ÌèâÍ∑†ÏùÄ Ï§ëÍ∞ÑÏãúÌóò ÏÑ±Ï†ÅÏùò ÌèâÍ∑†Í≥º Í∞ôÍ±∞ÎÇò Îçî ÎÇÆÎã§) ŒºÍ∏∞Îßê‚â§ŒºÏ§ëÍ∞Ñ
# H1: (Í∏∞ÎßêÏãúÌóò ÏÑ±Ï†ÅÏùò ÌèâÍ∑†ÏùÄ Ï§ëÍ∞ÑÏãúÌóò ÏÑ±Ï†ÅÏùò ÌèâÍ∑†Î≥¥Îã§ Îçî ÎÜíÎã§) ŒºÍ∏∞Îßê>ŒºÏ§ëÍ∞Ñ

import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt

data= pd.read_excel("C:/Users/darkr/OneDrive/Desktop/2ÌïôÎÖÑ1ÌïôÍ∏∞/ÌôïÌÜµ/Î∞∞ÌÑ∞Î¶¨,Í∑ºÎ°ú,ÏÑ±Ï†Å,Ï≤¥Ï§ë/ÌôïÎ•†Î∞èÌÜµÍ≥Ñ_ÏÑ±Ï†Å.xlsx")
print(data[['midterm','final']].describe())
data.boxplot(column=['midterm','final'], vert=False)
plt.show()

result=stats.ttest_rel(data.midterm, data.final, alternative='greater')
print(result)                          
```
