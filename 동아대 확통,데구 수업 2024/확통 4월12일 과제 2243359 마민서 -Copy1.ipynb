```python
# 3  

import numpy as np
from matplotlib import pyplot as plt 
from scipy.stats import binom

plt.rcParams['font.family'] ='D2coding'

n=3
p=0.5

x= np.arange(n+1) 
prob= binom.pmf(x, n, p) 


plt.bar(x, prob)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title(' 3 (n=3, p=0.5)')
plt.show()
print(prob)
```

```python
#   

import numpy as np
from matplotlib import pyplot as plt 
from scipy.stats import binom

plt.rcParams['font.family'] ='D2coding'

x= np.array([1, 2, 3, 4, 5])
count= np.array([8, 12, 15, 85, 30])
prob= count/150
plt.bar(x,prob)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title(' (150)')
plt.show()
print(np.round(prob,3))
```
