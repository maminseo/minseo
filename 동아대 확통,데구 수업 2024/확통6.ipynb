```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

data = pd.DataFrame(pd.read_excel("C:/Users/darkr/OneDrive/Desktop/2학년1학기/확통/외식비.xlsx"))

colors = ['yellow', 'blue', 'red', 'green', 'orange', 'pink', 'black', 'brown']


for i,column in enumerate(data.columns[1:]):
    plt.hist(data[column], label ='bins=5', bins=5, color = colors[i])
    plt.xlabel(data)
    plt.ylabel('weight')
    plt.legend()
    plt.show()


print("왜도: ", stats.skew(data))
```
