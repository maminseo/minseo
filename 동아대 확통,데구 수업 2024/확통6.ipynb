```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

data = pd.DataFrame(pd.read_excel("C:/Users/darkr/OneDrive/Desktop/2학년1학기/확통/외식비.xlsx"))
colors = ['yellow', 'blue', 'red', 'green', 'orange', 'pink', 'black', 'brown']
data.set_index("지역", inplace=True)

for i,col in enumerate(data.columns):
    plt.hist(data[col], label ='bins=5', bins=5, color = colors[i])
    plt.xlabel(data.columns[i])
    plt.ylabel('지역별 외식비')
    plt.legend()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

data = pd.DataFrame(pd.read_excel("C:/Users/darkr/OneDrive/Desktop/2학년1학기/확통/외식비.xlsx"))

colors = ['yellow', 'blue', 'red', 'green', 'orange', 'pink', 'black', 'brown']


for i,column in enumerate(data.columns[1:]):
    plt.hist(data[column], label ='bins=5', bins=5, color = colors[i])
    plt.xlabel(data)
    plt.ylabel('weight')
    plt.legend()
    plt.show()


print("왜도: ", stats.skew(data))
```
