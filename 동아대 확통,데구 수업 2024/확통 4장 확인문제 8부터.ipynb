```python
# 확인문제 8

import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
plt.rcParams['font.family'] ='D2coding'

mu1 = 7
sigma1 = 15

x = np.arange(mu1-70, mu1+70, 0.1)
y = norm.pdf(x, mu1, sigma1) 

plt.bar(x, y)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('정규분포그래프(1)(mu=7, sigma=15)')
plt.show()

mu2 = 175
sigma2 = 20

x = np.arange(mu2-70, mu2+70, 0.1)
y = norm.pdf(x, mu2, sigma2) 

plt.bar(x, y, color = 'red')
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('정규분포그래프(2)(mu=175, sigma=20)')
plt.show()
```

### 확인문제 10 - 표로 계산

1. P(0<=Z<=1.5) 0.43319
2. P(1<=Z<=1.5) 0.09185
3. P(-1.26<=Z<=1.26) 0.89617-(1-0.89617) = 0.89617 - 0.10383 = 0.79234
4. P(-2.5<=Z<=-1.5) 0.06681 - 0.00621 = 0.0606
5. P(Z<=-1.5) 0.06681
6. P(2<=Z) 1-0.97725 = 0.02275

파이썬 계산과 동일하지만 미세한 오차가 발생함.

```python
# 확인문제 10 - 파이썬으로 계산

from scipy.stats import norm 

mu=0
sigma=1

#1
y1 = norm.cdf(1.5, mu, sigma) 
y2 = norm.cdf(0, mu, sigma)
print('P(0≤Z≤1.5)=', y1-y2)

#2
y1 = norm.cdf(1.5, mu, sigma)
y2 = norm.cdf(1, mu, sigma)
print('P(1≤Z≤1.5)=', y1-y2)

#3
y1 = norm.cdf(1.26, mu, sigma)
y2 = norm.cdf(-1.26, mu, sigma)
print('P(-1.26≤Z≤1.26)=', y1-y2)

#4
y1 = norm.cdf(-1.5, mu, sigma)
y2 = norm.cdf(-2.5, mu, sigma)
print('P(-2.5≤Z≤-1.5)=', y1-y2)

#5
y1 = norm.cdf(-1.5, mu, sigma)
print('P(Z≤-1.5)=', y1)

#6
y1 = norm.cdf(2, mu, sigma)
print('P(2≤Z)=', 1-y1)
```

```python
#예시 검색한거

from scipy.stats import norm

# (1) P(0 ≤ Z ≤ 1.5)
prob1 = norm.cdf(1.5) - norm.cdf(0)

# (2) P(1 ≤ Z ≤ 1.5)
prob2 = norm.cdf(1.5) - norm.cdf(1)

# (3) P(-1.26 ≤ Z ≤ 1.26)
prob3 = norm.cdf(1.26) - norm.cdf(-1.26)

# (4) P(-2.5 ≤ Z ≤ -1.5)
prob4 = norm.cdf(-1.5) - norm.cdf(-2.5)

# (5) P(Z ≤ -1.5)
prob5 = norm.cdf(-1.5)

# (6) P(2 ≤ Z)
prob6 = 1 - norm.cdf(2)

# 결과 출력
print("(1) P(0 ≤ Z ≤ 1.5):", prob1)
print("(2) P(1 ≤ Z ≤ 1.5):", prob2)
print("(3) P(-1.26 ≤ Z ≤ 1.26):", prob3)
print("(4) P(-2.5 ≤ Z ≤ -1.5):", prob4)
print("(5) P(Z ≤ -1.5):", prob5)
print("(6) P(2 ≤ Z):", prob6)
```

### 확인문제 13 - 표를 이용해서 계산

1. P(Z≤k)=0.95, k=1.64
2. P(Z≤k)=0.975, k=1.96
3. P(Z≤k)=0.05, k=-1.65
4. P(Z≤k)=0.01, k=-2.33
5. P(Z≤k)=0.9, k=1.28

파이썬과 값이 동일하지만 미세한 오차있음

```python
# 확인문제 13 - 파이썬 이용

from scipy.stats import norm

mu=0
sigma=1

#1
percent_point=0.95 
print('P(Z≤k)=0.95, k=',norm.ppf(percent_point, mu, sigma))

#2
percent_point=0.975
print('P(Z≤k)=0.975, k=',norm.ppf(percent_point, mu, sigma))

#3
percent_point=0.05 
print('P(Z≤k)=0.05, k=',norm.ppf(percent_point, mu, sigma))

#4
percent_point=0.01
print('P(Z≤k)=0.01, k=',norm.ppf(percent_point, mu, sigma))

#5
percent_point=0.9
print('P(Z≤k)=0.9, k=',norm.ppf(percent_point, mu, sigma))
```
