```python
#불량품 개수의 대한 확률뷴표

import numpy as np
import matplotlib.pyplot as plt

x= np.array([0, 1, 2, 3, 4])
count= np.array([5, 25, 40, 25, 5])
prob= count/100
plt.bar(x,prob)
plt.show()
```

```python
#동전 2번 던지기 막대그래프

import matplotlib.pyplot as plt

arr= [0, 1, 2]
parr= [0.25, 0.5, 0.25]
plt.bar(arr, parr)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.show()
```

```python
#동전 5번 던지기

import numpy as np
from matplotlib import pyplot as plt 
from scipy.stats import binom

n=5
p=0.5

# 0~n까지의 리스트 생성(확률변수)
x= np.arange(n+1) # n+1=6 따라서 1,2,3,4,5 까지 가능 
mean, var= binom.stats(n, p)

# x의 각 확률을 list로 생성하여 prob에 저장
prob= binom.pmf(x, n, p) 

print('mean=', mean, 'var=', var)
plt.bar(x, prob)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('binomial distribution(n=5, p=0.5)')
plt.show()
print(prob)
```

```python
#4장 문제 3번 

import numpy as np
from matplotlib import pyplot as plt 
from scipy import stats

n=5
p=0.5


x= np.arange(n+1) 
mean, var= stats.binom.stats(n, p)


prob= stats.binom.pmf(x, n, p) 

print('mean=', mean, 'var=', var)
plt.bar(x, prob)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('binomial distribution(n=5, p=0.5)')
plt.show()
print(prob)

print(stats.binom.pmf(2, n, p)) #문제2
```

```python
#4장 문제 4번 : 주사위 3의 배수 (주사위 3번 던짐)

import numpy as np
from matplotlib import pyplot as plt 
from scipy.stats import binom

plt.rcParams['font.family'] ='D2coding'

n=3
p=1/3

x= np.arange(n+1) 
mean, var= binom.stats(n, p)

prob= binom.pmf(x, n, p) 

print('mean=', mean, 'var=', var)
plt.bar(x, prob)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('주사위 3의 배수(n=2, p=1/3)')
plt.show()
print(prob)
```

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import binom

n=5
p=0.5

x = np.arange(n+1)
cprob= binom.cdf(x, n, p)

plt.bar(x, cprob)
plt.xlabel('X')
plt.ylabel('P(X<=x)')
plt.show()
print(cprob)
print()

# 3회이하일확률만별도로출력하고싶을때
print('3회이하일확률=', binom.cdf(3, n, p))
```

```python
#4장 35p

import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm

mu = 50
sigma = 5

x = np.arange(mu-30, mu+30, 0.1)
y = norm.pdf(x, mu, sigma) 

plt.bar(x, y)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('normal distribution(mu=50, sigma=5)')
plt.show()
```

```python
#중간고사때 문제를 코딩으로 분석: 수학 A영역 평균53,표준편차20 / 수학 B영역 평균57,표준편차15 정규분표 그래프 
#둘리는 A영역 75점 - Z스코어: (75-53) / 20 = 1.1 
#마루는 B영역 75점 - Z스코어: (75-57) / 15 = 1.2 마루 승!
# 그럼 각자 상위 몇프로인가? 표준정규분포표를 보거나 파이썬으로 알아볼수있음

import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
plt.rcParams['font.family'] ='D2coding'

muA = 53
sigmaA = 20

x = np.arange(0, 100, 0.1)
y = norm.pdf(x, muA, sigmaA) 

plt.bar(x, y)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('수학A영역(mu=53, sigma=20)')
plt.show()

muB = 57
sigmaB = 15

x = np.arange(0, 100, 0.1)
y = norm.pdf(x, muB, sigmaB) 

plt.bar(x, y, color = 'yellow')
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('수학B영역(mu=57, sigma=15)')
plt.show()
```
