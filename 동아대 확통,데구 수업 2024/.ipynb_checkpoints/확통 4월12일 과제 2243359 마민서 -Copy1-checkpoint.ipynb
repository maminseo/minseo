#### 1. 확률변수 X를 앞면이 나오는 회수로 정했을 때, 동전을 3번 던졌을시에 확률분포표를 작성하세요.

동전을 던질 때 앞면이 나올 확률 = p , 뒷면이 나올 확률 = 1−p. 동전의 던지기가 공평하다고 가정하면 p = 0.5 <br>
X가 각각의 값에 대해 나타내는 확률은 이항 확률분포를 따름. 동전을 3번 던진다고 가정하면, X는 0, 1, 2, 3 중 하나의 값을 가질 수 있다.


| 사건 | 확률변수(X) | 확률 P(X) |
|---|:---:|:---|
| 세 번의 던지기 중 아무 앞면도 나오지 않을 확률 | X = 0 | 0.125 |
| 세 번의 던지기 중 딱 한 번 앞면이 나올 확률 | X = 1 | 0.375 |
| 세 번의 던지기 중 두 번 앞면이 나올 확률 | X = 2 | 0.375 |
| 세 번의 던지기 중 모두 앞면이 나올 확률 | X = 3 | 0.125 |





p(X) 식 직접 계산


```python
#동전 3번 던지기 막대그래프

import numpy as np
from matplotlib import pyplot as plt 
from scipy.stats import binom

plt.rcParams['font.family'] ='D2coding'

n=3
p=0.5

x= np.arange(n+1) 
prob= binom.pmf(x, n, p) 


plt.bar(x, prob)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('동전 3번 던지기(n=3, p=0.5)')
plt.show()
print(prob)
```

#### 2. 영화관 별점 확률분포

영화를 150일동안 개봉했다. 확률변수 X를 별점의 개수로 정했을때 빈도수 세보니 다음과 같았다. 확률분포표를 작성하시오

| 별점(X) | 도수 | (상대도수)확률 |
|---|:---:|:---:|
| 5개 | 30 | 0.2 |
| 4개 | 85 | 0.567 |
| 3개 | 15 | 0.1 |
| 2개 | 12 | 0.08 |
| 1개 | 8 | 0.053 |
| 합계 | 150 | 1 |다.

```python
# 영화관 별점 막대그래프

import numpy as np
from matplotlib import pyplot as plt 
from scipy.stats import binom

plt.rcParams['font.family'] ='D2coding'

x= np.array([1, 2, 3, 4, 5])
count= np.array([8, 12, 15, 85, 30])
prob= count/150
plt.bar(x,prob)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('영화관 별점(150명)')
plt.show()
print(np.round(prob,3))
```
